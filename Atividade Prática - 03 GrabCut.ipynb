Exercício 1

Faça uma pesquisa na internet de uma imagem contendo um veículo com uma paisagem no fundo.
Você pode escolher a imagem da sua preferência.
Não esqueça que é necessário ter uma paisagem  no fundo.

Aplique o algoritmo de segmentação GrabCut para realizar a separação entre o plano de fundo (background) e o plano da frente (foreground). Demonstre os resultados obtidos.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img1 = cv.imread('carro.jpg')
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)

#Retângulo de Segmentação
img2 = img1.copy()
p1 = (100,50)           # canto superior esquerdo do retângulo
p2 = (600,350)          # canto inferior direito do retângulo
img2 = cv.rectangle(img2,p1,p2,(255,0,0),2)

#Parâmetros
mascara = np.zeros(img1.shape[:2], np.uint8)
bgModel = np.zeros((1,65), np.float64)
fgModel = np.zeros((1,65), np.float64)
retangulo = p1+p2

#Aplicar o algoritmo GrabCut
cv.grabCut(img1, mascara, retangulo, bgModel, fgModel, 5, cv.GC_INIT_WITH_RECT)

#Separar o Background do Foreground
#Após o algoritmo executar, a máscara contém valores entre 0 e 3.
#Os valores 0 e 2 representam o Background
#Os valores 1 e 3 representam o Foreground

filtro = np.where ( (mascara == 0) | (mascara == 2),0,1).astype('uint8')
img3 = img1.copy()
img3 = img3*filtro[:,:,np.newaxis]

plt.figure(figsize=(25,25))
plt.subplot(131), plt.imshow(img1)
plt.subplot(132), plt.imshow(img2)
plt.subplot(133), plt.imshow(img3)

```
